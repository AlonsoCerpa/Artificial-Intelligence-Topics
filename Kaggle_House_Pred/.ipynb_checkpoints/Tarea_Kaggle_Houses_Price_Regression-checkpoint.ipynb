## 1. Lectura y exploración de datos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df_train = pd.read_csv('train.csv')
y_train = df_train['SalePrice']
df_train.drop(['SalePrice'],axis=1,inplace=True)
df_test = pd.read_csv('test.csv')
id_test = df_test['Id']
```

```python
y_train.shape
```

```python
num_rows_train = df_train.shape[0]
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
df = pd.concat([df_train, df_test], sort=False)
```

```python
df.shape
```

```python
df.drop(['Id'],axis=1,inplace=True)
```

```python
#eliminar columnas si el numero de nulos en la columna es mayor a la mitad del numero de filas
half_rows = df.shape[0] // 2
for col in df.columns:
    number_of_na_in_col = df[col].isna().sum()
    if number_of_na_in_col > half_rows:
        df.drop([col], axis=1, inplace=True)
    elif number_of_na_in_col > 0:
        if df[col].dtype == object or df[col].dtype == np.int64:
            df[col] = df[col].fillna(df[col].mode()[0])
        elif df[col].dtype == np.float64:
            df[col] = df[col].fillna(df[col].mean())
            
```

```python
df.shape
```

```python
df.isnull().sum().sum()
```

```python
#codificar variables categoricas en booleanas con 'get_dummies'
obj_cols = list(df.select_dtypes(include=['object']).columns)
df = pd.get_dummies(df, columns=obj_cols)
```

```python
df.shape
```

## 2. Predicción con distintos modelos

### 2.1 Regresión lineal multivariada propia

```python
import linear_regression as lr
```

```python
X = df
```

```python
X_norm, X_mean_data, X_standard_dev = lr.Normalizar_Datos(X)
print(X_norm)
```

```python
X_norm.shape
```

```python
X_train_norm = X_norm.iloc[:num_rows_train,:]
X_test_norm = X_norm.iloc[num_rows_train:,:]
```

```python
X_train_norm.shape
```

```python
X_test_norm.shape
```

```python
y_train_norm, y_train_mean_data, y_train_standard_dev = lr.Normalizar_Datos(y_train)
print(y_train_norm)
```

```python
X_train_norm = np.c_[X_train_norm, np.ones(X_train_norm.shape[0])]        #bias
```

```python
X_train_norm
```

```python
X_test_norm = np.c_[X_test_norm, np.ones(X_test_norm.shape[0])]        #bias
```

```python
X_test_norm
```

```python
y_train_norm = y_train_norm.values
```

```python
y_train_norm
```

```python
learn_rate = 0.01
num_iter = 10000
W_gradient_desc = lr.Crear_Pesos(X_train_norm)
W_gradient_desc, costs = lr.Gradiente_Descendiente(X_train_norm, y_train_norm, W_gradient_desc, num_iter, learn_rate)
```

```python
W_gradient_desc
```

```python
cost_train_gradient_desc = lr.Calcular_Costo(X_train_norm, y_train_norm, W_gradient_desc)
print(cost_train_gradient_desc)
```

```python
plt.plot(range(len(costs)), costs)
```

```python
X_test_norm.shape
```

```python
W_gradient_desc.shape
```

```python
y_pred = (np.matmul(X_test_norm, W_gradient_desc) * y_train_standard_dev) + y_train_mean_data
```

```python
y_pred.shape
```

```python
y_pred
```

```python
id_test.shape
```

```python
id_test = id_test.astype(int)
```

```python
id_test
```

```python
submission_data_dict = {'Id': id_test, 'SalePrice': y_pred}
submission_data = pd.DataFrame(data=submission_data_dict)
```

```python
submission_data
```

```python
submission_data.to_csv("sample_submission.csv", index=False)
```
