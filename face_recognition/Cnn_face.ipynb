```python
import torch
import torchvision
import numpy as np
from torchvision import datasets
import torchvision.transforms as transforms
from torch.utils.data.sampler import SubsetRandomSampler
```

```python
np.random.seed(2)
torch.manual_seed(0)

batch_size = 32
#porcentaje de pruebas
test_size = 0.3
#porcentaje de validacion
valid_size = 0.1

#Composicion de transformaciones a las imagenes de entrada
transform = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(20),
    transforms.Resize(size=(224,224)),
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))
    ])
```

```python
#Cargador de datos genérico para imágenes
data = datasets.ImageFolder('caras', transform=transform)
print(len(data))
```

```python
#Division del conjunto de Datos en: Entrenamiento (70%) y Pruebas (30%)
num_data = len(data)
indices_data = list(range(num_data))
np.random.shuffle(indices_data)
split_tt = int(np.floor(test_size * num_data))
train_idx, test_idx = indices_data[split_tt:], indices_data[:split_tt]

#Division del conjunto de Entrenamiento en: Entrenamiento (63% del total de Datos) y Validación (7% del total de Datos)   
num_train = len(train_idx)
split_tv = int(np.floor(valid_size * num_train))
train_new_idx, valid_idx = train_idx[split_tv:], train_idx[:split_tv]

#Definimos objetos para extraer muestras de los datos
train_sampler = SubsetRandomSampler(train_new_idx)
test_sampler = SubsetRandomSampler(test_idx)
valid_sampler = SubsetRandomSampler(valid_idx)
```

```python
train_loader = torch.utils.data.DataLoader(data, batch_size=batch_size,
    sampler=train_sampler, num_workers=1)
valid_loader = torch.utils.data.DataLoader(data, batch_size=batch_size, 
    sampler=valid_sampler, num_workers=1)
test_loader = torch.utils.data.DataLoader(data, batch_size=batch_size, 
    sampler=test_sampler, num_workers=1)
classes = ('alonso', 'anabel', 'david', 'elard', 'harry', 'rodrigo')
```

```python
print("Número de caras para el entrenamiento =", len(train_sampler))
print("Número de caras para la validación =", len(valid_sampler))
print("Número de caras para las pruebas =", len(test_sampler))
```

```python
print("Tamaño del batch =", batch_size)
print()
print("Número de batchs para el entrenamiento =", len(train_loader))
print("Número de batchs para la validación =", len(valid_loader))
print("Número de batchs para las pruebas =", len(test_loader))
```

```python
print("Para los datos de entrenamiento:\n")
for i, batch in enumerate(train_loader):
    print("Tamaño de la entrada del batch %d ="%(i), batch[0].size())
print()
    
for i, batch in enumerate(train_loader):
    print("Tamaño del ground truth del batch %d ="%(i), batch[1].size())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

#Función para denormalizar y mostrar la imagen
def imshow(img):
    img = img / 2 + 0.5  #Denormalizar
    plt.imshow(np.transpose(img, (1, 2, 0)))  #Convertir de Tensor a Imagen
    
#Extrayendo un batch de 32 imágenes y sus etiquetas del conjunto de Entrenamiento
dataiter = iter(train_loader)
images, labels = dataiter.next()

imshow(torchvision.utils.make_grid(images))

print(' '.join('%5s' % classes[labels[j]] for j in range(32)))
```

```python
import torch.nn as nn
import torch.nn.functional as F

#Arquitectura de la CNN
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.conv1 = nn.Conv2d(3, 6, 5) #Imagen de entrada de 3 canales, salida de 6 canales
        self.pool = nn.MaxPool2d(2, 2)
        self.conv2 = nn.Conv2d(6, 32, 5) #Imagen de entrada de 6 canales, salida de 32 canales
        self.fc1 = nn.Linear(32*53*53, 120) #Imagen de entrada de 53x53 con 32 canales (convertida a vector) 
        self.fc2 = nn.Linear(120, 84)
        self.fc3 = nn.Linear(84, len(classes))

    def forward(self, x):
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = x.view(-1, 32*53*53)
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x


net = Net()
```

```python
import torch.optim as optim

#Especifica la funcion perdida
criterion = torch.nn.CrossEntropyLoss()

#Especifica el optimizador
optimizer = torch.optim.SGD(net.parameters(), lr=0.003, momentum=0.9)
```

```python
#Número de épocas para entrenar
n_epochs = 15

valid_loss_min = np.Inf

for epoch in range(1, n_epochs+1):

    train_loss = 0.0
    valid_loss = 0.0
    
    #Red en modo entrenamiento
    net.train()
    for data, target in train_loader:
    
        #Limpiar las gradientes de los parámetros del modelo
        optimizer.zero_grad()
        #Calcular la predicción en base a la entrada (Propagación hacia adelante)
        output = net(data)
        #Calcular el valor de la función de pérdida
        loss = criterion(output, target)
        #Calcular la gradiente de la pérdida con respecto a los parámetros del modelo (Backpropagation)
        loss.backward()
        #Realizar un paso de la optimización = Actualizar los parámetros del modelo
        optimizer.step()
        #Actualizar la pérdida del entrenamiento
        train_loss += loss.item()*data.size(0)
            
    #Red en modo evaluación/pruebas
    net.eval()
    for data, target in valid_loader:
        #Propagación hacia adelante
        output = net(data)
        #Calcular el valor de la función de pérdida
        loss = criterion(output, target)
        #Actualizar la pérdida de la validación
        valid_loss += loss.item()*data.size(0)
    
    # calculate average losses
    train_loss = train_loss/len(train_loader.dataset)
    valid_loss = valid_loss/len(valid_loader.dataset)
        
    # imprimir training/validacion 
    print('Época: {} \tPérdida de entrenamiento: {:.6f} \tPérdida de validación: {:.6f}'.format(
        epoch, train_loss, valid_loss))
```

```python
dataiter = iter(test_loader)
images, labels = dataiter.next()

# print images
imshow(torchvision.utils.make_grid(images))
print('GroundTruth: ', ' '.join('%5s' % classes[labels[j]] for j in range(32)))
```

```python
outputs = net(images)
```

```python
_, predicted = torch.max(outputs, 1)
print('Predicted: ', ' '.join('%5s' % classes[predicted[j]]
                              for j in range(32)))
```

```python
test_loss = 0.0
class_correct = list(0. for i in range(len(classes)))
class_total = list(0. for i in range(len(classes)))

net.eval()

for data, target in test_loader:
    
    #Propagación hacia adelante
    output = net(data)
    #Calcular la pérdida
    loss = criterion(output, target)
    #Actualizar la pérdida de las pruebas
    test_loss += loss.item()*data.size(0)
    
    _, pred = torch.max(output, 1)
    
    correct_tensor = pred.eq(target.data.view_as(pred))
    correct = np.squeeze(correct_tensor.numpy())
    
    for i in range(len(target)):
        label = target.data[i]
        class_correct[label] += correct[i].item()
        class_total[label] += 1

#Pérdida de pruebas promedio
test_loss = test_loss/len(test_loader.dataset)
print('Pérdida de pruebas: {:.6f}\n'.format(test_loss))

for i in range(len(classes)):
    if class_total[i] > 0:
        print('Prueba de %5s: %2d%% (%2d/%2d)' % (
            classes[i], 100 * class_correct[i] / class_total[i],
            np.sum(class_correct[i]), np.sum(class_total[i])))
    else:
        print('Clase vacia')


```
